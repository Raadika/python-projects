```python
#1) What is the average price of cars across different brands, and which brands have the highest and lowest average prices?
avg_price_by_brand = data.groupby("Brand")["Price"].mean()
```

```python
avg_price_by_brand = avg_price_by_brand.sort_values(ascending=False)
```

```python
plt.bar(avg_price_by_brand.index, avg_price_by_brand.values)
```

```python
#2) How does the condition of a car (Status) affect its price?
plt.boxplot(
    [data[data["Status"] == status]["Price"] for status in data["Status"].unique()],
    labels=data["Status"].unique(),
    notch=True,)
```
