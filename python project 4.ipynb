```python
country_confirmed_cases.plot(kind='bar')
```

```python
#2) Which countries or regions have the highest and lowest total confirmed cases and fatalities?
highest_confirmed_country = data.loc[data['ConfirmedCases'] == data['ConfirmedCases'].max(), 'Country_Region'].values[0]
```

```python
lowest_confirmed_country = data.loc[data['ConfirmedCases'] == data['ConfirmedCases'].min(), 'Country_Region'].values[0]
```

```python
highest_fatalities_country = data.loc[data['Fatalities'] == data['Fatalities'].max(), 'Country_Region'].values[0]
```

```python
lowest_fatalities_country = data.loc[data['Fatalities'] == data['Fatalities'].min(), 'Country_Region'].values[0]
```

```python
results_df = pd.DataFrame({
    'Metric': ['Highest Confirmed', 'Lowest Confirmed', 'Highest Fatalities', 'Lowest Fatalities'],
    'Country/Region': [highest_confirmed_country, lowest_confirmed_country, highest_fatalities_country, lowest_fatalities_country]
})
```

```python

plt.bar(results_df['Metric'], results_df['Country/Region'])
```
